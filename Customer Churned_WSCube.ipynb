- Converting 0 and 1 in the SeniorCitizen Column to Yes and No.

```python
def convert(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
    
df["SeniorCitizen"] = df["SeniorCitizen"].apply(convert)
df
```

```python
ax= sns.countplot(x = "Churn", data= df, color="#af9659")

ax.bar_label(ax.containers[0])

plt.title("Count of Customers Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
gb
```

```python
plt.figure(figsize= (3,3))
plt.pie(gb["Churn"], labels = gb.index, autopct = "%1.2f%%", colors=["#add60d","#fbc840"])
plt.title("Percentage of Customers Churned", fontsize= 14)
plt.show()
```

- From the given pie chart we can conclude that 26.54% of our customers have churned out.
* Now let's explore the reasons behind it.

```python
plt.figure(figsize= (5,3))
sns.countplot(x= "gender", data= df, hue="Churn", palette= "viridis")

plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize= (4,3))
sns.countplot(x= "SeniorCitizen", data= df, hue="Churn", palette= "rainbow")

plt.title("Churn by Senior Citizen")
plt.show()
```

```python

# Assuming df is your DataFrame
plt.figure(figsize=(9,6))

# Calculate percentage values
grouped = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize=True).unstack() * 100

# Plot stacked bar chart
grouped.plot(kind="bar", stacked=True, colormap="cividis", figsize=(4,3))

# Add percentage labels
for i, total in enumerate(grouped.sum(axis=1)):
    for j, val in enumerate(grouped.iloc[i]):
        plt.text(i, grouped.iloc[i, :j+1].sum() - val/2, f"{val:.1f}%", ha="center", fontsize=10, color="white")

plt.title("Churn by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", bbox_to_anchor=(1,1))
plt.xticks(rotation=0)
plt.show()
```

- Comparitive a greater percentage of people in senior citizen category have churned.

```python
plt.figure(figsize=(20,6))
sns.histplot(x= "tenure", data=df, bins=72, hue="Churn")

plt.show()
```

- People who have used company services for a long time have stayed and people who have used services for 1 2 months have churned.

```python
plt.figure(figsize= (5,4))
ax= sns.countplot(x= "Contract", data= df, hue="Churn", palette= "rainbow")
for container in ax.containers:
    ax.bar_label(container)
plt.title("Count of Customers by Contract")
plt.show()
```

- People who have month to month contract are likely to churn than from those who have 1 or 2 years of contract.

```python

# Sample DataFrame (Replace this with actual data)
# df = pd.read_csv("your_data.csv") 

# Define the columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
           'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
           'PaperlessBilling', 'PaymentMethod', 'Churn']

# Set up subplots
num_cols = 3  # Number of columns in subplot grid
num_rows = -(-len(columns) // num_cols)  # Ceiling division to determine rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 4 * num_rows))
axes = axes.flatten()  # Flatten in case of a single row

# Loop through columns and create countplots
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i], palette="viridis", hue=df["Churn"])
    axes[i].set_title(f"Countplot of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=45)  # Rotate labels if needed

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

- The countplots illustrate the distribution of various categorical features in the dataset, segmented by churn status. Key insights include a higher churn rate among customers with fiber optic internet, month-to-month contracts, and electronic check payments. Features like online security, tech support, and longer contracts appear to correlate with lower churn. The majority of customers have phone service, and paperless billing is prevalent among both churned and retained users.
